```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/insurance.csv")
```

```python
df.head()
```

```python
male_data = df[df['sex'] == 'male']
female_data = df[df['sex'] == 'female']
regions = df['region'].unique()
region_wise_data = {region: df[df['region'] == region] for region in regions}

print("Male Data:")
print(male_data.head())

print("\nFemale Data:")
print(female_data.head())

print("\nRegion-wise Data:")
for region, data in region_wise_data.items():
    print(f"\nRegion: {region}")
    print(data.head())
```

```python
ages = df['age']
print("\nVector (Ages):")
print(ages.head())
```

```python
smoker_factor = df['smoker'].astype('category')
print("\nFactor (Smoker):")
print(smoker_factor.head())
print("Categories:", smoker_factor.cat.categories)
```

```python
my_list = [ages.head().tolist(), smoker_factor.head().tolist()]
print("\nList:")
print(my_list)
```

```python
matrix = df[['age', 'expenses']].head().to_numpy()
print("\nMatrix (Age and Expenses):")
print(matrix)
```

```python
subset_df = df[['age', 'sex', 'smoker', 'region', 'expenses']].head()
print("\nData Frame:")
print(subset_df)
```

```python
import matplotlib.pyplot as plt

smoker_expenses = df[df['smoker'] == 'yes']['expenses'].mean()
non_smoker_expenses = df[df['smoker'] == 'no']['expenses'].mean()

plt.bar(['Smoker', 'Non-Smoker'], [smoker_expenses, non_smoker_expenses])
plt.title('Average Expenses: Smoker vs. Non-Smoker')
plt.xlabel('Smoker Status')
plt.ylabel('Average Expenses')
plt.show()
```

```python
plt.scatter(df['age'], df['expenses'])
plt.title('Age vs. Expenses')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.show()
```

```python
smokers_gender_count = df[df['smoker'] == 'yes']['sex'].value_counts()
plt.figure(figsize=(8, 6))
smokers_gender_count.plot(kind='bar', color=['blue', 'pink'])
plt.title('Number of Smokers Gender-wise', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
filtered_df = df[(df['age'] >= 20) & (df['age'] <= 40)]
smoker_counts = filtered_df.groupby('age')['smoker'].value_counts().unstack().fillna(0)

smoker_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Smoker Status by Age (20-40)')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.xticks(rotation=45)
plt.legend(title='Smoker', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
teen_smokers = df[(df['age'] < 20) & (df['smoker'] == 'yes')]
num_teen_smokers = len(teen_smokers)
print(f"The number of teenage smokers is: {num_teen_smokers}")
```
